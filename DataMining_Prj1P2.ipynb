```python
df = pd.read_csv('/content/drive/My Drive/masters_portal.csv')
print(df.shape)
df = df.dropna(subset=['university_name', 'program_name'])
print(df.shape)
df.head() 
```

### Convert the whole dataframe to lower case

```python
df = df.apply(lambda x: x.astype(str).str.lower())
df.head()
```

```python
columns_to_filter = ['structure', 'academic_req', 'facts', 'city']
```

```python
import string
PUNCT_TO_REMOVE = string.punctuation
def remove_punctuation(text):
    """custom function to remove the punctuation"""
    return text.translate(str.maketrans('', '', PUNCT_TO_REMOVE))

for item in columns_to_filter:
  df[item] = df[item].apply(lambda text: remove_punctuation(text))
  # df[item] = df[item].apply(lambda x: x.astype(str).str.replace(r'[!"#$%&\'()*+,-./:;<=>?@\\^_{|}~`+]', ''))
  # df[item] = df[item].apply(lambda x: x.astype(str).str.replace('[', ''))
  # df[item] = df[item].apply(lambda x: x.astype(str).str.replace(']', ''))
df.head()
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

STOPWORDS = set(stopwords.words('english'))
def remove_stopwords(text):
    """custom function to remove the stopwords"""
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])

for item in columns_to_filter:
  df[item] = df[item].apply(lambda text: remove_stopwords(text))
df.head()
```

```python
from collections import Counter
def remove_freqwords(text, freqWords):
    """custom function to remove the frequent words"""
    return " ".join([word for word in str(text).split() if word not in freqWords])

def remove_rarewords(text, rareWords):
    """custom function to remove the rare words"""
    return " ".join([word for word in str(text).split() if word not in rareWords])


columns_to_filter = ['structure', 'academic_req', 'facts']
for item in columns_to_filter:
  cnt = Counter()
  for text in df[item].values:
    for word in text.split():
      cnt[word] += 1
  freqWords = set([key for key in cnt.keys() if cnt[key] > 30000])
  if len(freqWords) < 10:
    freqWords = set([w for (w, wc) in cnt.most_common(10)])
  print('item: ', item, ' freqWords: ', len(freqWords), end=' ')
  df[item] = df[item].apply(lambda text: remove_freqwords(text, freqWords))

  rareWords = set([key for key in cnt.keys() if cnt[key] < 5])
  if len(rareWords) < 10:
    rareWords = set([w for (w, wc) in cnt.most_common()[:-10-1:-1]])
  print('rareWords: ', len(rareWords))
  df[item] = df[item].apply(lambda text: remove_rarewords(text, rareWords))

df.head()
```

```python
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()
def stem_words(text):
    return " ".join([stemmer.stem(word) for word in text.split()])
for item in columns_to_filter:
  df[item] = df[item].apply(lambda text: stem_words(text))
df.head()
```

```python
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')

lemmatizer = WordNetLemmatizer()
wordnet_map = {"N":wordnet.NOUN, "V":wordnet.VERB, "J":wordnet.ADJ, "R":wordnet.ADV}
def lemmatize_words(text):
    pos_tagged_text = nltk.pos_tag(text.split())
    return " ".join([lemmatizer.lemmatize(word, wordnet_map.get(pos[0], wordnet.NOUN)) for word, pos in pos_tagged_text])

for item in columns_to_filter:
  df[item] = df[item].apply(lambda text: lemmatize_words(text))
df.head()
```
